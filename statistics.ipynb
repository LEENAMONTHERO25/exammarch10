# 1.What is the meaning of six sigma in statistics?  Give proper example


Six Sigma is a statistical methodology used in quality management and process improvement to reduce defects or errors to a level that meets or exceeds customer expectations. It aims to improve the quality of processes by identifying and removing the causes of defects and minimizing variability.

The term "Six Sigma" comes from statistics and refers to the goal of achieving a process performance level where the probability of producing a defect is extremely low, typically defined as no more than 3.4 defects per million opportunities (DPMO). This corresponds to achieving a process capability level of at least 6σ (sigma), where σ represents the standard deviation of the process.

Here's an example to illustrate the concept of Six Sigma:

Suppose a manufacturing process produces widgets, and the specification for the length of each widget is 100 ± 2 millimeters. However, due to variability in the process, some widgets may be longer or shorter than the target length, leading to defects.

To achieve Six Sigma performance, the process should be improved to ensure that almost all widgets fall within the specified range with minimal variation. This means reducing the standard deviation of the process to the point where the probability of producing a widget outside the specified range is less than 3.4 defects per million.

For example, if the process standard deviation is reduced to 0.33 millimeters, which is one-sixth of the specification tolerance (2 millimeters), then the process capability would be 6σ. This would result in very few defects, as almost all widgets would fall within the specified range of 100 ± 2 millimeters.

# What type of data does not have a log-normal distribution or a Gaussian distribution?  Give proper example


Skewed data: Skewed data distributions are asymmetric, with a longer tail on one side of the distribution compared to the other. Skewed data can be positively skewed (right-skewed) or negatively skewed (left-skewed). An example of positively skewed data is household income, where a few individuals may have very high incomes, causing the distribution to be skewed to the right. Conversely, the distribution of waiting times at a busy restaurant may be negatively skewed, as most people would not want to wait for too long.

Bimodal or multimodal data: Bimodal or multimodal distributions have two or more distinct peaks or modes. These distributions indicate the presence of multiple underlying processes or populations within the data. An example of bimodal data is the distribution of body temperatures, which may have one mode for healthy individuals and another mode for individuals with a fever.

Heavy-tailed distributions: Heavy-tailed distributions have tails that decay more slowly than the tails of a Gaussian distribution. These distributions are characterized by a higher probability of extreme values compared to a Gaussian distribution. Examples include the distribution of stock returns, where extreme positive or negative returns occur more frequently than expected under a normal distribution.

Discrete distributions: Discrete distributions represent data that can only take on specific, distinct values. Unlike continuous distributions, which can take on any value within a range, discrete distributions have gaps between possible values. Examples include the distribution of the number of goals scored in a soccer game or the distribution of the number of children in a family.

# What is the meaning of the five-number summary in Statistics? Give proper example

The five-number summary is a set of descriptive statistics that provides a concise summary of the distribution of a dataset. It consists of five values: the minimum, the first quartile (Q1), the median (Q2), the third quartile (Q3), and the maximum. These values help to understand the central tendency, spread, and shape of the data distribution.

Here's what each value represents:

Minimum: The smallest value in the dataset.
First Quartile (Q1): The value below which 25% of the data falls. Also known as the lower quartile.
Median (Q2): The middle value of the dataset when it is ordered from least to greatest. It represents the 50th percentile of the data.
Third Quartile (Q3): The value below which 75% of the data falls. Also known as the upper quartile.
Maximum: The largest value in the dataset.
The five-number summary is often depicted graphically using a boxplot or box-and-whisker plot, where a box is drawn from Q1 to Q3 with a line indicating the median. The "whiskers" extend from the minimum to the maximum, providing additional information about the spread of the data and identifying potential outliers.

Here's an example to illustrate the five-number summary:

Consider the following dataset representing the scores of 20 students on a math test:

65,72,75,78,80,82,85,87,88,90,92,94,95,96,97,98,99,100,100,100

The five-number summary for this dataset would be:

Minimum: 65
Q1 (First Quartile): 80 (the median of the first half of the dataset)
Median: 89 (the middle value of the dataset)
Q3 (Third Quartile): 96 (the median of the second half of the dataset)
Maximum: 100

# What is correlation? Give an example with a dataset & graphical representation on jupyter Notebook



Correlation is a statistical measure that describes the strength and direction of a relationship between two variables. It indicates how much and in what direction two variables change together. A correlation value ranges from -1 to 1:

A correlation coefficient of 1 indicates a perfect positive correlation, meaning that as one variable increases, the other variable also increases in a linear fashion.
A correlation coefficient of -1 indicates a perfect negative correlation, meaning that as one variable increases, the other variable decreases in a linear fashion.
A correlation coefficient of 0 indicates no linear relationship between the variables.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate example data
np.random.seed(0)
x = np.random.normal(loc=0, scale=1, size=100)  # Variable 1
y = 2 * x + np.random.normal(loc=0, scale=1, size=100)  # Variable 2 (linearly related to Variable 1)

# Create DataFrame
df = pd.DataFrame({'Variable 1': x, 'Variable 2': y})

# Plot the scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Variable 1', y='Variable 2')
plt.title('Scatter Plot of Variable 1 vs. Variable 2')
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.grid(True)
plt.show()
```

```python
# Calculate correlation coefficient
correlation_coefficient = df['Variable 1'].corr(df['Variable 2'])
print("Correlation coefficient between Variable 1 and Variable 2:", correlation_coefficient)

# Plot the scatter plot with correlation coefficient
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Variable 1', y='Variable 2')
plt.title('Scatter Plot of Variable 1 vs. Variable 2')
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.grid(True)
plt.text(x=-2, y=5, s=f'Correlation coefficient = {correlation_coefficient:.2f}', fontsize=12, color='red')
plt.show()
```
